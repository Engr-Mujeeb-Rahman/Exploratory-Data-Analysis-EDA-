#### Q. 2) Remove all the records where the Confirmed Cases is Less Than 1000

```python
remove_Confirmed_Cases=Data[Data["Confirmed"]<1000]
remove_Confirmed_Cases
```

```python
# Now remove this from the dataset
listindex=list()
for i in remove_Confirmed_Cases.index:
    listindex.append(i)

listindex
    
```

```python
Data=Data.drop(listindex)
Data
```

 Here you can see that we can sucssfilly remove the data

```python
Data[Data["Confirmed"]<1000]
```

#### Q. 3) In which Date, maximum number of Confirmed cases were recorded ?

```python
Data.groupby("Date")["Confirmed"].sum().sort_values(ascending=False)
```
