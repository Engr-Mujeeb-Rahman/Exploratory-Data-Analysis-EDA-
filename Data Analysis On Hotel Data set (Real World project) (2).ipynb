# Data anlysis and Visulization

```python
percentage_of_cancelled=df["is_canceled"].value_counts()
percentage_of_cancelled
```

```python
percentage_of_cancelled=df["is_canceled"].value_counts(normalize=True)
percentage_of_cancelled
```

```python
df["is_canceled"].unique()
```

```python
plt.figure(figsize=(6,5))
plt.bar(["NotConceled","canceled"],percentage_of_cancelled,edgecolor='k',width=0.7)
plt.title("Reservation status")
plt.show()
```

```python
# Now checked that which hotel has more cancaltion rate
sns.countplot(x=df["hotel"],data=df,hue=df["is_canceled"],palette="Blues",width=0.7)
plt.xlabel("Hotel")
plt.ylabel("Resevation")
plt.show()
```
